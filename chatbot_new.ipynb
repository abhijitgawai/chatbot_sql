```python
import sqlite3

# Connect to a new database file
conn = sqlite3.connect('Chinook.db')

# Open the SQL script file and read its contents
with open('./Chinook_Sqlite.sql', encoding="utf8") as f:
    sql_script = f.read()
# print(sql_script)
# Execute the SQL script in the new database
conn.executescript(sql_script)
# print(sql_script)

# Close the database connection
conn.close()
```

```python
import sqlite3

# Connect to an existing database file
conn = sqlite3.connect('Chinook.db')

# Create a cursor object to execute SQL queries
cursor = conn.cursor()

# Execute a SELECT query
cursor.execute('SELECT * FROM PlaylistTrack limit 10')

# Fetch the query results
rows = cursor.fetchall()

# Print the results
for row in rows:
    print(row)

# Close the cursor and database connection
cursor.close()
conn.close()
```
